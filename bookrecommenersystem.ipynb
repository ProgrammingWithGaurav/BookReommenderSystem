## Popularity Based Recommendation System

```python
ratings.merge(books,on='ISBN').shape
```

```python
ratings_with_name = ratings.merge(books ,on='ISBN')
```

```python
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df.head()
```

```python
ratings_with_name
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'}, inplace=True)
avg_rating_df.head()
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df.shape
```

```python
popular_df2 = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df2 =popular_df2.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M', 'num_ratings', 'avg_rating']]
```

## Collabrative Filtering Based Recomender System

```python
ratings_with_name
```

```python
# Users who have rated more than 200 books
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
max_rating_users = x[x].index
```

```python
max_rating_users
```

```python
filtered_rating=  ratings_with_name[ratings_with_name['User-ID'].isin(max_rating_users)]
```

```python
# Books with more than 50 ratings by users
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score= cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]
    
    data = []
    # looping items
    for i in similar_items:
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item = temp_df.drop_duplicates('Book-Title')
        data.append(item)
    return data
```

```python
recommend('''Harry Potter and the Chamber of Secrets (Book 2)''')[0]
```
